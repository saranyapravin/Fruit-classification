```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import glob
from skimage import io, color, transform
import csv
import pandas as pd
import imutils
features=[["S.No","Area","width","Height"]]
```

```python
color_feature=[["R_value","G_value","B_value"]]
def color_feature_extract(cimage):
    average_color_per_row = np.average(cimage, axis=0)
    average_color = np.average(average_color_per_row, axis=0)
    average_color = np.uint8(average_color)
    color_feature.append(average_color)
    return color_feature
```

```python
def resizeAndPad(img, size, padColor=0):

    h, w = img.shape[:2]
    sh, sw = size

    # interpolation method
    if h > sh or w > sw: # shrinking image
        interp = cv2.INTER_AREA
    else: # stretching image
        interp = cv2.INTER_CUBIC

    # aspect ratio of image
    aspect = w/h  # if on Python 2, you might need to cast as a float: float(w)/h

    # compute scaling and pad sizing
    if aspect > 1: # horizontal image
        new_w = sw
        new_h = np.round(new_w/aspect).astype(int)
        pad_vert = (sh-new_h)/2
        pad_top, pad_bot = np.floor(pad_vert).astype(int), np.ceil(pad_vert).astype(int)
        pad_left, pad_right = 0, 0
    elif aspect < 1: # vertical image
        new_h = sh
        new_w = np.round(new_h*aspect).astype(int)
        pad_horz = (sw-new_w)/2
        pad_left, pad_right = np.floor(pad_horz).astype(int), np.ceil(pad_horz).astype(int)
        pad_top, pad_bot = 0, 0
    else: # square image
        new_h, new_w = sh, sw
        pad_left, pad_right, pad_top, pad_bot = 10, 10, 10, 10

    # set pad color
    if len(img.shape) is 3 and not isinstance(padColor, (list, tuple, np.ndarray)): # color image but only one color provided
        padColor = 255

    # scale and pad
    scaled_img = cv2.resize(img, (new_w, new_h), interpolation=interp)
    scaled_img = cv2.copyMakeBorder(scaled_img, pad_top, pad_bot, pad_left, pad_right, borderType=cv2.BORDER_CONSTANT, value=(255,255,255))

    return scaled_img

#v_img = cv2.imread('v.jpg') # vertical image
#scaled_v_img = resizeAndPad(v_img, (200,200), 127)

#h_img = cv2.imread('h.jpg') # horizontal image
#scaled_h_img = resizeAndPad(h_img, (200,200), 127)
```

```python
filename ="D:\\SARANYA-PhD\\Dataset\\Training_set\\Apple2\\*.jpg"
i=0
for im in glob.glob(filename):
    i=i+1
    image = cv2.imread(im)
    image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
    r_image = resizeAndPad(image, (200,200), 127)
    
    blurred=cv2.pyrMeanShiftFiltering(r_image,15,60)
    #plt.imshow(blurred)
    #plt.show()
    image1 = cv2.cvtColor(blurred, cv2.COLOR_RGB2GRAY)
    #plt.imshow(image1,'gray')
    #plt.show()
    ret, imgf = cv2.threshold(image1, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
    print(ret)
    #plt.imshow(imgf,'gray')
    #plt.show()
    struct_elem = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))
    smooth_bin_img = cv2.erode(imgf, struct_elem, iterations=1)
    smooth_bin_img = cv2.dilate(smooth_bin_img, struct_elem, iterations=2)
    #plt.imshow(smooth_bin_img,'gray')
    #plt.show()
    _, contours,_ = cv2.findContours(smooth_bin_img, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    print('number of contours',len(contours))
    cv2.drawContours(r_image,contours,1,(255,0,0),2)
    #plt.imshow(r_image)
    #plt.show()
    color_feature=color_feature_extract(r_image)
    area=cv2.contourArea(contours[1])
    print(area)
    x,y,w,h=cv2.boundingRect(contours[1])
    img=cv2.rectangle(r_image,(x,y),(x+w,y+h),(0,255,0),2)
    #plt.imshow(r_image)
    #plt.show()
    features.append([i,area,w,h])
print(features)
path ="C:/Users/SUN TEX/Desktop/Features/APPLE_MYFEATURES.csv"
with open(path, "w") as output:
    writer = csv.writer(output,lineterminator='\n')
    writer.writerows(features)
print("write completed")
path ="C:/Users/SUN TEX/Desktop/Features/APPLE_color_features.csv"
with open(path, "w") as output:
    writer = csv.writer(output,lineterminator='\n')
    writer.writerows(color_feature)
print("write completed")
```
